```python
import folium
from geopy.geocoders import Nominatim

# Getting the coordinates of Lisbon
geolocator = Nominatim(user_agent="geoapiExercises")
location = geolocator.geocode("Lisbon, Portugal")
lisbon_lat, lisbon_long = location.latitude, location.longitude

# Creating a map centered at Lisbon
lisbon_map = folium.Map(location=[lisbon_lat, lisbon_long], zoom_start=10)

# Adding cities to the map
cities = ['Alcântara', 'Campo de Ourique', 'Misericórdia',
       'Santa Maria Maior', 'Ajuda', 'Arroios', 'Campolide', 'Benfica',
       'Beato', 'Belém', 'Areeiro', 'Marvila', 'Avenidas Novas',
       'Estrela', 'Parque das Nações', 'Santa Clara', 'Olivais',
       'Santo António', 'Lumiar', 'São Vicente', 'Carnide',
       'Penha de França', 'Alvalade', 'São Domingos de Benfica']
for city in cities:
    city_location = geolocator.geocode(city)
    if city_location:
        folium.Marker(location=[city_location.latitude, city_location.longitude], popup=city).add_to(lisbon_map)
lisbon_map
```
